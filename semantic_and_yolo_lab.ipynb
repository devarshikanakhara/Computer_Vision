#Practical 3: Semantic Segmentation using U-Net

```python
# =============================================================================
# Practical 3: Semantic Segmentation using U-Net (Auto-Load Data Version)
# =============================================================================

import tensorflow as tf
from tensorflow.keras import layers, models, Input
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds  # Library to load datasets automatically

# ────────────────────────────────────────────────
#   Step 1: Define U-Net model
# ────────────────────────────────────────────────
def double_conv_block(x, n_filters):
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    return x

def downsample_block(x, n_filters):
    f = double_conv_block(x, n_filters)
    p = layers.MaxPool2D(2)(f)
    return f, p

def upsample_block(x, conv_features, n_filters):
    x = layers.Conv2DTranspose(n_filters, 3, 2, padding="same")(x)
    x = layers.concatenate([x, conv_features])
    x = double_conv_block(x, n_filters)
    return x

def build_unet(input_shape=(128, 128, 1)):
    inputs = Input(input_shape)

    # Encoder
    f1, p1 = downsample_block(inputs, 64)
    f2, p2 = downsample_block(p1, 128)
    f3, p3 = downsample_block(p2, 256)
    f4, p4 = downsample_block(p3, 512)

    # Bottleneck
    bottleneck = double_conv_block(p4, 1024)

    # Decoder
    u6 = upsample_block(bottleneck, f4, 512)
    u7 = upsample_block(u6, f3, 256)
    u8 = upsample_block(u7, f2, 128)
    u9 = upsample_block(u8, f1, 64)

    # Output
    outputs = layers.Conv2D(1, 1, padding="same", activation="sigmoid")(u9)
    return models.Model(inputs, outputs, name="U-Net")

def dice_loss(y_true, y_pred):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return 1 - (2. * intersection + 1) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + 1)

# ────────────────────────────────────────────────
#   Step 2: Load Data via Code (Fixed Version 4.0.0)
# ────────────────────────────────────────────────
def load_and_preprocess_data():
    print("Downloading/Loading Oxford-IIIT Pet dataset...")

    # FIX: Use version 4.0.0 to match Google Colab's available data
    dataset, info = tfds.load('oxford_iiit_pet:4.0.0', with_info=True)

    def preprocess(data_item):
        # 1. Resize Image and Mask to 128x128
        image = tf.image.resize(data_item['image'], (128, 128))
        mask = tf.image.resize(data_item['segmentation_mask'], (128, 128))

        # 2. Convert Image to Grayscale (to match input_shape=(128,128,1))
        image = tf.image.rgb_to_grayscale(image)

        # 3. Normalize Image [0,1]
        image = image / 255.0

        # 4. Preprocess Mask
        # The dataset uses: 1=Pet, 2=Background, 3=Border
        # We assume 1 is the target (Pet) and everything else is background (0)
        mask = tf.where(mask == 1, 1.0, 0.0)

        return image, mask

    # Create train and validation batches
    train_ds = dataset['train'].map(preprocess).batch(16).prefetch(tf.data.AUTOTUNE)
    val_ds = dataset['test'].map(preprocess).batch(16).prefetch(tf.data.AUTOTUNE)

    return train_ds, val_ds

# ────────────────────────────────────────────────
#   Step 3: Execution
# ────────────────────────────────────────────────
if __name__ == "__main__":
    # 1. Load Data
    train_dataset, val_dataset = load_and_preprocess_data()
    print("Data ready.")

    # 2. Build Model
    model = build_unet(input_shape=(128, 128, 1))
    model.summary()

    # 3. Compile
    model.compile(optimizer=Adam(learning_rate=1e-4),
                  loss=dice_loss,
                  metrics=['accuracy'])

    # 4. Train
    print("Starting Training...")
    history = model.fit(train_dataset,
                        validation_data=val_dataset,
                        epochs=5,  # 5 epochs is enough for a quick demo
                        verbose=1)

    # 5. Visualize Results
    print("Visualizing Prediction...")
    for image_batch, mask_batch in val_dataset.take(1):
        pred_batch = model.predict(image_batch)

        plt.figure(figsize=(12, 4))

        # Input
        plt.subplot(1, 3, 1)
        plt.imshow(image_batch[0, :, :, 0], cmap='gray')
        plt.title("Input Image")

        # Ground Truth
        plt.subplot(1, 3, 2)
        plt.imshow(mask_batch[0, :, :, 0], cmap='gray')
        plt.title("Ground Truth Mask")

        # Prediction
        plt.subplot(1, 3, 3)
        plt.imshow(pred_batch[0, :, :, 0] > 0.5, cmap='gray')
        plt.title("Predicted Mask")

        plt.show()
        break # Only show one batch

    print("Practical 3 completed successfully!")
```


# Practical 3: Mini U-Net for Single Image Tumor Detection



```python

import tensorflow as tf
from tensorflow.keras import layers, models, Input
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

# ────────────────────────────────────────────────
#   Step 1: Define "Mini" U-Net (Modified Architecture)
# ────────────────────────────────────────────────
def conv_block(x, n_filters):
    # Simplified block: Conv2D -> ReLU (Removed BatchNormalization for single image stability)
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu")(x)
    x = layers.Conv2D(n_filters, 3, padding="same", activation="relu")(x)
    return x

def build_mini_unet(input_shape=(128, 128, 1)):
    inputs = Input(input_shape)

    # 1. Encoder (Downsampling) - Only 2 Levels
    # Level 1
    f1 = conv_block(inputs, 32)
    p1 = layers.MaxPool2D(2)(f1)

    # Level 2
    f2 = conv_block(p1, 64)
    p2 = layers.MaxPool2D(2)(f2)

    # 2. Bottleneck
    bottleneck = conv_block(p2, 128)

    # 3. Decoder (Upsampling)
    # Level 2 Up
    u2 = layers.Conv2DTranspose(64, 3, 2, padding="same")(bottleneck)
    u2 = layers.concatenate([u2, f2])
    u2 = conv_block(u2, 64)

    # Level 1 Up
    u1 = layers.Conv2DTranspose(32, 3, 2, padding="same")(u2)
    u1 = layers.concatenate([u1, f1])
    u1 = conv_block(u1, 32)

    # Output
    outputs = layers.Conv2D(1, 1, padding="same", activation="sigmoid")(u1)

    return models.Model(inputs, outputs, name="Mini-U-Net")

def dice_loss(y_true, y_pred):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return 1 - (2. * intersection + 1) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + 1)

# ────────────────────────────────────────────────
#   Step 2: Load Image (Same as before)
# ────────────────────────────────────────────────
def load_single_image(image_path):
    if not os.path.exists(image_path):
        print(f"Error: {image_path} not found. Please upload it!")
        return None, None

    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (128, 128))

    # Auto-mask thresholding
    _, mask = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

    # Normalize
    img = img / 255.0
    img = np.expand_dims(img, axis=-1)
    img = np.expand_dims(img, axis=0)

    mask = mask / 255.0
    mask = np.expand_dims(mask, axis=-1)
    mask = np.expand_dims(mask, axis=0)

    return img, mask

# ────────────────────────────────────────────────
#   Step 3: Execution
# ────────────────────────────────────────────────
if __name__ == "__main__":
    IMAGE_FILE = "my_image.jpg" # <--- Ensure your file is named this

    print(f"Loading {IMAGE_FILE}...")
    x_train, y_train = load_single_image(IMAGE_FILE)

    if x_train is not None:
        print("Building Mini U-Net...")
        # USE THE NEW MINI MODEL
        model = build_mini_unet(input_shape=(128, 128, 1))

        model.compile(optimizer=Adam(learning_rate=0.001), # Slightly higher LR helps single image
                      loss=dice_loss,
                      metrics=['accuracy'])

        print("Training (100 epochs to ensure convergence)...")
        # Increased epochs because we really want it to memorize this image
        history = model.fit(x_train, y_train,
                            epochs=100,
                            verbose=0) # verbose=0 to reduce clutter, prints 'Done' after
        print("Training Done.")

        # Visualize
        print("Predicting...")
        pred = model.predict(x_train)

        plt.figure(figsize=(12, 4))
        plt.subplot(1, 3, 1); plt.imshow(x_train[0,:,:,0], cmap='gray'); plt.title("Original Image")
        plt.subplot(1, 3, 2); plt.imshow(y_train[0,:,:,0], cmap='gray'); plt.title("Target Mask")
        plt.subplot(1, 3, 3); plt.imshow(pred[0,:,:,0] > 0.5, cmap='gray'); plt.title("Mini U-Net Prediction")
        plt.show()
```

#Practical 4: Real-time Object Detection with YOLOv8

```python
# =============================================================================
# Practical 4: Save YOLOv8 Detection to Video File
# =============================================================================
from ultralytics import YOLO
import cv2
import os

# 1. Load Model
model = YOLO("yolov8n.pt")
print("Model loaded ✓")

# 2. Setup Input Video
input_file = "video.mp4"

if not os.path.exists(input_file):
    print(f"Error: '{input_file}' not found. Please upload it!")
else:
    cap = cv2.VideoCapture(input_file)

    # 3. Get Video Properties (Width, Height, FPS)
    # We need these to ensure the output video matches the input
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = cap.get(cv2.CAP_PROP_FPS)

    # 4. Setup Output Video Writer
    # We will save as 'output.mp4' using the 'mp4v' codec
    output_file = "output.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_file, fourcc, fps, (width, height))

    print(f"Processing video... Saving to '{output_file}'")

    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break # End of video

        # Run YOLOv8 inference
        results = model(frame, verbose=False)

        # Draw boxes (returns a numpy array compatible with OpenCV)
        annotated_frame = results[0].plot()

        # Write the processed frame to the output video file
        out.write(annotated_frame)

        # Progress tracker
        frame_count += 1
        if frame_count % 20 == 0:
            print(f"Processed {frame_count} frames...")

    # 5. Cleanup
    cap.release()
    out.release()
    print("------------------------------------------------")
    print(f"Done! Video saved as '{output_file}'")
    print("Click the folder icon on the left to download it.")
```
